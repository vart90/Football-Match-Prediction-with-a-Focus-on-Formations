```python
import requests
from bs4 import BeautifulSoup
import time
import csv

url = "https://fbref.com/en/comps/9/2022-2023/schedule/2022-2023-Premier-League-Scores-and-Fixtures"


time.sleep(10)
response = requests.get(url)

soup = BeautifulSoup(response.content, "html.parser")
def match_href_pattern(href):
    return href and href.startswith("/en/matches")

links = soup.find_all("a", attrs={"href": match_href_pattern})
print(links)
```

```python
for i in range(0,len(links)):
  fixture_url = "https://fbref.com" + links[i]['href']
  time.sleep(5)
  response = requests.get(fixture_url)
  fixture_html = response.content
  fixture_soup = BeautifulSoup(fixture_html, 'html.parser')
  form = fixture_soup.find_all("th", attrs={"colspan": 2})
  if form != []:
    form1 = form
    print(form1)
```

```python
import pandas as pd
import re


team_names = []
formations = []

for row in df1['COL1']:
    row_str = str(row)

    team_name_match = re.findall(r'>(.*?) \(', row_str)
    formation_match = re.findall(r'\((.*?)\)', row_str)

    team_names.extend(team_name_match)
    formations.extend(formation_match)

print(team_names)
print(formations)
```

```python
import pandas as pd


for i in range(len(formations)):
    if len(formations[i]) == 5:
        formations[i] += "-0"

hform = []
aform = []

team1 = []
team2 = []

for i in range(len(team_names)):
    if i % 2 == 0:
        hform.append(formations[i])
        team1.append(team_names[i])
    else:
        aform.append(formations[i])
        team2.append(team_names[i])


df = pd.DataFrame({'HTeam': team1, "ATeam":team2, 'Hformation':hform,'Aformation':aform})

df.to_csv('22-23_formations.csv', index=False)
```
